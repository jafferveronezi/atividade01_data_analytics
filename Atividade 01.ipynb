10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
engine = create_engine('sqlite:///dados/sqlite-sakila.db')
insp = inspect(engine)
```

```python
insp.get_table_names()
```

```python
teste = insp.get_table_names()

teste
```

```python
sql_query ='''
SELECT 
    * 
    FROM 
        city 
            WHERE city LIKE 'b%'    
 '''
insp_city_df = pd.read_sql_query(sql_query, engine)
insp_city_df.head()
```

```python
sql_query ='''
SELECT 
    * 
    FROM 
        country
            WHERE country LIKE 'b%'    
 '''
insp_country_df = pd.read_sql_query(sql_query, engine)
insp_country_df.head()
```

```python
sql_query ='''
SELECT 
    * 
    FROM 
        city, country
            WHERE country LIKE 'b%'    
 '''
insp_df = pd.read_sql_query(sql_query, engine)
insp_df.head()
```
